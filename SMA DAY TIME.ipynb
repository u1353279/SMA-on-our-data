#                                    # Performing SMA on our dataset FOR DAY TIME

```python
import pandas as pd
df=pd.read_excel('DAY.xlsx')
df.drop(df.columns[df.columns.str.contains('Unnamed',case=False)],axis=1, inplace=True)
df.head()
```

```python
import matplotlib.pyplot as plt
df["NO"].plot()
plt.figure(figsize=(10,10))
```

```python
moving_avg = df["NO"].rolling(15).mean()
plt.plot(df["NO"],color='black')
plt.plot(moving_avg, color='yellow')
plt.title("SMA FOR NO")
```

```python
moving_avg = df["1F"].rolling(15).mean()
plt.plot(df["1F"],color='black')
plt.plot(moving_avg, color='yellow')
plt.title("SMA FOR 1 Faulty")
```

```python
moving_avg = df["2F"].rolling(15).mean()
plt.plot(df["2F"],color='black')
plt.plot(moving_avg, color='yellow')
plt.title("SMA FOR 2 Faulty")
```
