#                                    # Performing SMA on our dataset FOR DAY TIME all faults

```python
import pandas as pd
df=pd.read_excel('DAY.xlsx')
df.drop(df.columns[df.columns.str.contains('Unnamed',case=False)],axis=1, inplace=True)
df.head()
```

```python
import matplotlib.pyplot as plt
df["NO"].plot()
plt.figure(figsize=(10,10))
```

```python
moving_avg = df["NO"].rolling(15).mean()
plt.plot(df["NO"],color='black')
plt.plot(moving_avg, color='yellow')
plt.title("SMA FOR NO")
```
